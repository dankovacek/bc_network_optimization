```python
import pandas as pd
import numpy as np

import os
import time

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()
```

```python
X = np.load('processed_data/concurrent_record_matrix.npy')
X.shape
```

```python
years = list(range(101))
complete_months = []
for n_years in years:
    min_record_met = np.where(X >= 12 * n_years, 1, 0)
    col_sums = np.nansum(min_record_met, axis=0)
    valid_stns = sum(np.where(col_sums > 0, 1, 0))
    complete_months.append(valid_stns)
    
```

```python
p = figure(width=600, height=400, tools='box_zoom,reset,save')

p.line(years, complete_months, color='green', line_width=3)
p.xaxis.axis_label = "Minimum concurrent record [years]"
p.yaxis.axis_label = "Num unique stations"
show(p)
```
