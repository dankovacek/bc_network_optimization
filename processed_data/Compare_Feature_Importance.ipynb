```python
import pandas as pd
import numpy as np
import os

from bokeh.plotting import figure, show, output_notebook
from bokeh.layouts import gridplot, column, row
from bokeh.models import ColumnDataSource
from bokeh.palettes import Category20, Set2
import numpy as np
output_notebook()
```

```python
BASE = os.getcwd()
```

```python
feature_folder = os.path.join(BASE, 'feature_importance')
feature_files = os.listdir(feature_folder)
```

```python
all_dfs = []
for f in feature_files:
    tdf = pd.read_csv(os.path.join(feature_folder, f))
    all_dfs.append(tdf)
```

```python
df = pd.concat(all_dfs)
df.drop(['Unnamed: 0', '0', '1', '2', '3'], inplace=True, axis=1)
```

```python
def create_histogram(data, feature, bins=30):
    # Compute histogram with numpy
    data = [e for e in data if np.isfinite(e)]
    hist, edges = np.histogram(data, bins=bins)

    # Create a new plot
    p = figure(title=feature, tools='', background_fill_color="#fafafa")
    p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)

    # Add labels and grid customization
    p.xgrid.grid_line_color=None
    p.y_range.start=0
    p.xaxis.axis_label = feature
    p.yaxis.axis_label = r'$$P(X)$$'
    p.grid.grid_line_color='gray'
    p.grid.grid_line_alpha=0.5
    # Return the Bokeh figure object
    return p
```

```python
feature_cols = [c for c in df.columns if ~c.endswith('_score')]
plots = []
for c in feature_cols:
    print(c)
    plot = create_histogram(df[c].values, c)
    plots.append(plot)
    
```

```python
layout = gridplot(plots, ncols=4, width=300, height=250)
show(layout)
```
